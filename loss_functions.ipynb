```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.1, 10, 10000)
plt.plot(x, x)
plt.plot(x, np.exp(np.sqrt(np.log(x)**2 + 1)))
plt.plot(x, np.log(x)**2)
plt.grid()
plt.show()
```
